```python
%%time
all_tweets = open('/kaggle/input/tweets/tweets_fr.txt')
N_TWEETS = 1e5
cleaned_tweets = []
tweets = []
targets_encoding = []
i = 0
for line in all_tweets:
    if i >= N_TWEETS: break
    
    current_emojis = emojis.get(line)
    
    encoded = np.array([emoji in current_emojis for emoji in top_100_empojis_1M_tweets], dtype=int)
    #print('sum:', encoded.sum())
    if encoded.sum() == 0 : 
        continue
    cleaned_tweet = ' '.join(re.sub("(@[A-Za-z0-9_]+)|([^0-9A-Za-zÃªÃ©Ã¨Ã Ã¹Ã¯Ã´Ã»Ã§Ã¢ \t])|(\w+:\/\/\S+)"," ",line).split())
    tweets.append(line)
    for emoji in current_emojis: 
        cleaned_tweet = cleaned_tweet.replace(emoji, '')
        cleaned_tweet = cleaned_tweet.strip()
    cleaned_tweets.append(cleaned_tweet)
    targets_encoding.append(encoded)
    i += 1 
all_tweets.close()
targets_encoding = np.array(targets_encoding)
targets_encoding.shape

     
```
